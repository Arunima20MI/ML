```python
import csv
import random
fn = "random_values.csv"
data = []
for i in range(5):
    row = [random.randint(1, 100) for i in range(5)]  
    data.append(row)
with open(fn, "w", newline="") as f:
    writer = csv.writer(f)
    header = ["Column1","Column2","Column3","Column4","Column5"]
    writer.writerow(header)
    writer.writerows(data)
print("Display the header file\n")
with open(fn, "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```
